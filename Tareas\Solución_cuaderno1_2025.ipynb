## Cálculo de pendiente y del offset (cero)

$$ y[t_k] \in [y_{min},y_{max}]$$

$$y[t_k] = m x[t_k] + c$$

$$m = \frac{y_{max}-y_{min}}{x_{max}-x_{min}}$$

$$c = y[t_k] - m x[t_k]$$

- Se crea un código tipo función bajo el ambiente def, para generar el cero y pendiente según los valores $y_{max}$ y $y_{min}$ definidos por el usuario.

```
y_min = 0 # valor min después de cero y pendiente
y_max = 5 # valor max después de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
```

```
yv = my_ceropen(xv,ymin=-15,ymax=20) #se llama la función definida en la celda anterior
```

```
''' Creamos la función que nos ayudará a ajustar la señal a un rango deseado organizando el cero y pendiente (cero_y_pendiente)'''

def cero_y_pendiente(signal_original, ymin=-7, ymax=7):
  # De la señal original, sacamos sus rangos (xmin, xmax)
  xmin = min(signal_original)
  xmax = max(signal_original)

  # Calculamos la pendiente
  m = (ymax-ymin)/(xmax-xmin)

  # Calculamos el offset escogiendo un valos específico (rango mínimo)
  c = ymin - m*xmin

  # Retornamos la señal ajustada
  return signal_original*m + c
```

# Creación de la señal original
Señal original senoidal con:
- Frecuencia de la señal $F_0 = 60 [Hz]$
- Amplitud de la señal $A = 120 [V]$
- Periodo de la señal $T_0 = \frac{1}{F_0}$

Como nuestra frecuencia de muestreo ($F_s$) debe ser al menos el doble que nuestra frecuencia máxima de la señal ($F_0$); se escogerá una frecuencia 100 veces mayor para mejorar la calidad de la representación digital.
- Frecuencia de muestreo $F_s = 100F_0$
- Periodo de muestreo $T_s = \frac{1}{F_s}$

```
''' Importamos las librería snecesarias '''
import numpy as np # numpy para las señales discretas
import matplotlib.pyplot as plt # librería para gráficas
from scipy.spatial.distance import cdist as dist # librería para calcular distancias

''' Creamos la señal con las variables anteriormente dichas '''
F_0 = 60 # Frecuencia de la señal [Hz]
A = 120 # Amplitud de la señal [V]
T_0 = 1/F_0 # Periodo de la señal [sg]
F_s = 100*F_0 # Frecuencia de muestreo [Hz]
T_s = 1/F_s # Periodo de muestreo [sg]

''' Creamos el vector de tiempo continuo junto con la señal seonidal '''
vt_continuo = np.linspace(0,2*T_0,int(2/T_s)) # Tiempo continuo para dos periodos

y = A*np.sin(2*np.pi*F_0*vt_continuo) # señal senoidal original
```

## Señal Discretizada

Es el resultado del proceso de muestreo. Aquí, la señal continua en el tiempo se convierte en una señal discreta en el tiempo al tomar muestras en intervalos regulares.

##### Proceso:
- Frecuencia de muestreo ($F_s$): La cual cunple con $F_s \geq 2F_0$
- Intervalo de muestreo ($T_s$): El intervalo de muestreo es $T_s = \frac{1}{F_s}$
- Muestreo: Se evalúa la señal en instantes de tiempo discretos $t_k=k \cdot T_s$, donde $k \in \mathbb{Z}$

```
vt_discreto = np.arange(0,2*T_0,T_s) # Tiempo discreto para dos periodos

y_discretizada = A*np.sin(2*np.pi*F_0*vt_discreto) # señal senoidal discretizada

# Realizamos el ajuste
y_discretizada = cero_y_pendiente(y_discretizada, ymin=0, ymax=10)

markerline, stemlines, baseline = plt.stem(vt_discreto, y_discretizada, 'r', label='Señal discretizada')
plt.setp(stemlines, linewidth=0.5)  # Ajusta el grosor de las líneas
plt.setp(markerline, markersize=4)  # (opcional) reduce el tamaño de los marcadores

plt.title("Señal discretizada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()
plt.show()
```

# Señal Cuantizada

Para un conversor de 3 bits hay $2^{3}=8$ niveles posibles

### Definimos los niveles de cuantización
El rango de la señal $[r_{min}, r_{max}]$, se divide en $2^b$ niveles, donde $\textbf{b}$ es el $\textbf{número de bits}$

- Se tienen una señal ajustada con valores en el rango [0,10].

### Asignar cada muestra al nivel más cercano
Esto implica calcular la distancia entre el valor de la señal y todos los niveles disponibles, y seleccionar el nivel con la menor distancia.

- Un ADC de 3 bits, lo que significa que hay $2^3=8$ niveles.

### Representar la señal cuantizada
Consta únicamente de los niveles seleccionados.

- Los niveles de cuantización los creamos con linspace, colocando nuestros rangos y dividinedo esos rangos en los niveles obtenidos.

```
# Definimos los niveles de cuantización:
r_min = 0
r_max = 10

# Definimos el número de niveles
bits = 3
nivel = 2**bits

# Representamos la señal creando el vector con los niveles de cuantización
V_cuantizacion = np.linspace(r_min, r_max, nivel)
#print(V_cuantizacion)

# Implementamos la cuantización utilizando cdist y argmin. Para un conversor de 3 bits. 8 niveles posibles.
def cuantizador(muestra, vector_cuantizado):
    """
    Cuantiza un valor yn al nivel más cercano en vq.

    Parámetros:
        muestra:            Valor individual de la señal ajustada a cuantizar.
        vector_cuantizado:  Vector de niveles de cuantización.

    Retorna:
        El nivel más cercano en el vector para el valor de la muestra.
    """
    # Convertimos la muestra a un array NumPy antes de usar reshape
    muestra = np.array(muestra)

    # Calculamos las distancias entre la muestra y todos los niveles del vector
    distancia_entre_niveles = cdist(muestra.reshape(-1, 1), vector_cuantizado.reshape(-1, 1))

    # Encontramos el índice del nivel más cercano
    nivel_mas_cercano = np.argmin(distancia_entre_niveles)

    '''print(f'Distancia entre niveles --> {distancia_entre_niveles}')
    print(f'Nivel más cercano --> {nivel_mas_cercano}')'''

    # Retornamos el valor del nivel más cercano en el vector de cuantización
    return vector_cuantizado[nivel_mas_cercano]

#cuantizador(3.2, V_cuantizacion)
```

# Cuantizamos y discretizamos

```
def discretizar_y_cuantizar(signal_original, F_s, bits, r_min, r_max):
  ''' DISCRETIZACIÓN '''
  T_s = 1/F_s # Definimos el intervalo de muestreo
  vt_discreto = np.arange(0,2*T_0,T_s) # Dedinimos el vector de tiempo discreto
  y_discreta = A*np.sin(2*np.pi*F_0*vt_discreto)

  #Ajustamos la señal
  y_discreta = cero_y_pendiente(y_discreta, ymin=0, ymax=10)

  ''' CUANTIZACIÓN '''
  nivel = 2**bits # Definimos los niveles de cuantización
  # Representamos la señal creando el vector con los niveles de cuantización
  V_cuantizacion = np.linspace(r_min, r_max, nivel)
  y_cuantizada = np.array([cuantizador(muestra, V_cuantizacion) for muestra in y_discreta])

  return vt_discreto, y_discreta, y_cuantizada
```

## Aplicando la función

```
# Vector de tiempo para dos periodos
vt_continuo = np.linspace(0,2*T_0,int(2/T_s)) # Tiempo continuo para dos periodos

y = A*np.sin(2*np.pi*F_0*vt_continuo) # señal senoidal original

# Ajuste de cero y pendiente
y = my_ceropen(y, ymin=0, ymax=10)

# Aplicar la función combinada
bits = 3
r_min = 0
r_max = 10
vt_discreto, y_discretizada, y_cuantizada = discretizar_y_cuantizar(y, F_s, bits, r_min, r_max)
```
